```python
#08 scalar product.py

F=np.array([2,7,3])
s = np.array([-2,3,4])

F_B = np.sqrt(np.dot(F,F))
s_B = np.sqrt(np.dot(s,s))

cos_Fs = np.dot(F,s)/(F_B*s_B)
angle=np.degrees(np.arccos(cos_Fs))
W = np.dot(F,s)

print("magnitude of force:",F_B,"N")
print("Magnitude of path:",s_B,"m")
print('Angle between F and s:',angle,"degrees")
print("Work",W,"Nm")
```

Dot Product = how much one vector projects onto another. dot = 0 then perpendicular  

Cross Product = a new vector that is perpendicular to the two input vectors

```python
# Cross Product 

F = np.array([2,7,-3])
l = np.array([2,3,4])

F_B = np.sqrt(np.dot(F,F))
l_B = np.sqrt(np.dot(l,l))

cos_Fl = np.dot(F,l)/(F_B*l_B)
angle=np.degrees(np.arccos(cos_Fl))

print("angle",angle,"degrees")

```

```python
# Dyadic Product - row vector * column vector 

A = np.array([1,2,3])
B = np.array([[4],[5],[6]])

C = np.outer(A,B)

print("Matrix A\n",A)
print("Matrix B\n",B)
print("Matrix C\n",C)
```

```python
# 3.3 Matrix Multiplication 

A = np.array([[2,3],[3,4]])
B = np.array([[5,6],[7,8]])

C = A@B
D = B@A

print("Matrix A\n",A)
print("Matrix B\n",B)
print("Matrix C\n",C)
print("Matrix D\n",D)

```
